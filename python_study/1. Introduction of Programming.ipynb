### 연산자의 종류

    - +, -, *, /, //, %, **
2. 

### + 산술 연산자의 우선순위


| 우선순위 | 연산자         | 연산                   |
|:----:|:-----------:|:--------------------:|
| 높음   | **          | 지수                   |
|      | -           | 음수 부호                |
|      | *, /, //, % | 곱셈, 나눗셈, 정수 나눗셈, 나머지 |
| 낮음   | +, -        | 덧셈, 뺄셈               |


```python
# 산술 연산자 우선순위 예시

print(-2 ** 4) 
## 지수(**) 먼저 -> 이후 음수(-) 표현 
# sol = -(2 ** 4) = -(16) = -16
print(-(2 ** 4))
# 지수(**) 먼저 -> 이후 음수(-) 표현
# sol = -(2 ** 4) = -(16) = -16
print((-2) ** 4)
# 지수(**) 먼저
# sol = (-2) ** 4 = 16
```

### 변수와 값, 메모리

메모리 주소(Memory Address) : 메모리 위치에 대한 식별자
- 메모리의 모든 위치에는 위치를 고유하게 식별할 수 있는 메모리 주소가 있다
- 컴퓨터 프로그램이나 하드웨어 장치가 데이터를 저장하고 나중에 이를 가져오는 장소
- 일반적으로 이진 형태의 숫자로 구성
객체(Object) : 타입을 갖는 메모리 주소 내의 값
- 메모리가 값에 할당되기 전까지는 객체가 존재하지 않음
- "값이 들어있는 상자"

### degrees = 36.5
변수 degrees에 값 36.5를 할당

- float type의 값 36.5가 메모리주소(491734)에 들어가있음 - 36.5는 객체가 됨
- 변수 degrees가 해당 객체 36.5를 참조
- 변수 degrees는 참조하는 객체 36.5의 메모리주소를 가짐

![image](https://github.com/ragu6963/TIL/assets/32388270/b5dfec1e-9f57-4dba-bf47-d333c988a1ab)


### 할당문

Variable = Expression 
1. 할당 연산자(=)의 오른쪽에 있는 표현식 평가
2. 평가 후 해당 값의 메모리주소 생성(표현식은 type과 메모리주소를 가진 객체가 됨)
3. 객체의 메모리 주소를 = 왼쪽에 있는 변수에 저장(변수가 해당 객체를 참조하고, 객체의 메모리 주소를 저장함)

할당문에서 사용한 변수가 존재하지 않는 변수일 경우 : 새 변수가 생성됨
할당문에서 사용한 변수가 기존에 존재했던 변수일 경우 : 재할당 과정 



### 변수의 재할당  
기존 변수를 재사용해서 변수에 들어있는 메모리 주소를 변경하는 것
- 변수가 어떤 객체를 참조하는지? 어떤 객체의 메모리주소를 저장하는지?



```python
# 변수의 재할당
number = 10 # number라는 변수가 10이라는 값(이제는 메모리주소가 생긴 객체)를 참조 
# 현재 number라는 변수에는 10의 메모리주소가 저장됨
double = 2 * number # double이라는 변수가 2 * number라는 값(객체)를 참조
# 현재 double이라는 변수에는 2 * number = 20의 메모리주소가 저장됨
print(double)
number = 5 # number라는 기존에 존재한 변수에 5라는 값을 재할당
# 현재 number라는 변수에는 5의 메모리주소가 저장됨
print(double) # double에는 아직 20의 메모리주소가 저장되어있음 
```

### Style Guide 
코드의 일관성과 가독성을 향상시키는 규칙과 권장 사항등의 모음
- [파이썬 스타일 가이드 PEP 8 참고 문서](https://peps.python.org/pep-0008/)
### 파이썬 Style Guide
1. 변수명 : 무엇을 위한 변수인지 직관적인 이름을 가져야 한다.
2. 들여쓰기 : 공백(space)를 4칸 사용하여 코드블럭을 들여쓴다.
3. 줄바꿈 : 한 줄의 길이는 79자로 제한하며, 길어질 경우 줄 바꿈을 사용한다.
4. Naming : 문자와 밑줄(_)을 사용하여 함수, 변수, 속성의 이름을 작성한다.
5. 함수와 클래스 : 함수, 클래스 등을 정의할 경우 블록 사이에 빈 줄을 추가한다.

### 주석

주석(Comment) : 프로그램 코드 내에 작성되는 설명이나 메모

### 주석의 목적
1. 코드의 특정부분을 설명하거나, 임시로 코드를 비활성화
2. 코드를 이해하거나 문서화
3. 다른 개발자에게 코드의 의도나 동작을 설명

```python
# 이것은 주석입니다.
print("위의 내용은 주석이라 인터프리터에 의해 나오지 않아요!")

'''
여러줄 주석은 이렇게 사용하면 됩니다!
'''

"""
이 방법도 있지요!
"""
```
