### 1. 암시적 형 변환

### 암시적 형 변환(Implicit Type Conversion)
파이썬이 자동으로 형 변환을 하는 것
- Boolean과 Num Type에서만 가능한 형 변환

```python
# 암시적 형 변환 예시
print(3 + 5.0) # int형 3을 float형 3.0으로 바꾸어 계산
print(True + 3) # Boolean형 True를 int형 1로 바꾸어 계산
print(True - False) # Boolean형 True와 False를 int형 1과 0으로 바꾸어 계산
```

### 2. 명시적 형 변환

### 명시적 형 변환(Explicit Type Conversion)
개발자가 직접 형 변환을 하는 것
- 암시적 형 변환이 아닌 모든 형 변환
- 특별 명시적 형 변환
    - str -> num : 형식에 맞는(str형의 데이터가 숫자로만 이루어진 경우) str 데이터에 한해 가능
    - num -> str : 모두 가능 

```python
# 명시적 형 변환 예시
print(str(3.141592))
print(int('1'))
print(float('3.141592'))
print(int(3.5))

# 주의할 점 : ValueError: invalid literal for int() with base 10: '3.5' - 파이썬 형 변환 에러
# print(int('3.5'))
# print(int(input('How tall are you? write cm.'))에서 소수형으로 적을 경우도 마찬가지
# 해결방안 : str -> float -> int로 형변환을 2번 진행
a = float('3.5')
print(int(a))

```
