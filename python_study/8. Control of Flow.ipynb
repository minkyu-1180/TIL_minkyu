### 1. 제어문

### 제어문(Control Statement)
코드의 실행 흐름을 제어하는 데 사용되는 구문
- 조건에 따라 코드블록을 실행하거나, 반복적으로 코드를 실행

### 2. 조건문

### 조건문(Conditional Statement)
주어진 조건식을 평가하여 해당 조건이 True인 경우에만 코드 블록을 실행하거나 건너뜀

### 파이썬 조건문에 사용되는 키워드
- if / elif / else

### if statement
if statement의 기본 구조



```python
# 조건문 예시
a = 5

if a > 3:
    print('3 초과')
else:
    print('3 이하')
```

### 복수 조건문
조건식을 순차적으로 비교하는 조건문(동시에 평가하는 것이 아님)

```python
# 복수 조건문 예시

dust = 30

if dust > 150:
    print('미세먼지 매우 많음')
elif dust > 80: # 150 >= dust > 80
    print('미세먼지 많음')
elif dust > 50 : # 80 >= dust > 50
    print('미세먼지 보통')
else: # 50 >= dust
    print('날씨 좋음')
```

### 중첩 조건문
조건문이 중첩되어 표현하는 것(if 안의 if)

```python
# 중첩 조건문 예시
dust = 500

if dust > 150:
    print('매우 나쁨')
    # 중첩 조건문
    if dust > 300:
        print('위험해요! 나가지 마세요!')
elif dust > 80:
    print('나쁨')
elif dust > 30:
    print('보통')
else:
    print('좋음')
```

### 3. 반복문

### 반복문(Loop Statement)
주어진 코드 블록을 여러 번 반복하여 실행하는 구문
- 특정 작업을 반복하여 수행(for)
- 주어진 조건이 참인 동안 반복하여 수행(while)
반복문에 사용되는 키워드 : for / while
### for statement
임의의 시퀀스 항목들을 그 시퀀스에 들어있는 순서대로 반복하는 것

### for statement의 기본 구조



### 반복 가능한 객체(iterable)
반복문에서 순회할 수 있는 객체
- 시퀀스 객체, dict, set, ,,,

# 추가내용 : iterable vs sequence
iterable : 객체의 원소를 하나씩 반환 가능하다.
- 순서가 정의될 필요 X
- 객체의 동작 방식을 정의
sequence : 원소가 순서대로 나열되어 있다. -> 인덱스가 존재한다
- 객체의 형태를 정의

sequence 형태는 아닌데 iterable한 객체 : dict, set
- 둘다 원소 순서는 없지만 

### For문의 원리
1. iter 객체 내의 첫 항목이 반복 변수에 할당되고, 코드블록을 실행한다.
2. 다음으로 반복 변수에 두 번째 항목이 할당되고, 코드블록을 실행한다.
3. 2번 과정을 반복하며 마지막으로 반복 변수의 iter 객체 내의 마지막 항목이 할당되고 코드블록이 실행된다.

```python
# for문 예제
items = ['apple', 'banana', 'candy', 'dancing']
for item in items: # iterable 객체인 items의 원소들을 반복변수 item에 차례대로 할당
    print(item) # 값이 할당된 변수와 관련된 값 출력
print()

chars = 'Korea'
for char in chars:
    print(char)
print()

new_dict = {
            'a' : 1,
            'b' : 2, 
            'c' : 3,
           }

for key in new_dict:
    print(f'{key} : {new_dict[key]}')
print()

new_set = {1, 2, 3, 4, 5}
for ele in new_set:
    print(ele)
print()

for i in range(1, 11):
    print(i)
```

### 인덱스로 리스트 순회
리스트의 인덱스로 접근하여 해당 요소들을 변경

```python
# 인덱스로 리스트 순회
numbers = [4, 6, 10, -8, 5]

for i in range(len(numbers)):
    numbers[i] = numbers[i] * 2
print(numbers)
```

```python
# 중첩 반복문
outers = ['A', 'B']
inners = ['c', 'd']

for outer in outers:
    for inner in inners:
        print(outer, inner)
```

### 중첩 리스트 순회
안쪽 리스트 요소에 접근하려면 바깥 리스트를 순회하면서 중첩 반복문을 사용해 각 안쪽 반복을 순회

```python
# 인덱스로 중첩 리스트 순회
elements = [['A', 'B'], ['C', 'D']]

for element in elements:
    print(element)
print()


for element in elements:
    for ele in element:
        print(ele)
```

### while statement
주어진 조건식이 True일 동안 코드를 반복해서 실행
- 조건식이 False가 될 때 까지 반복하는 것이라 생각하는 것이 편함

### while statement의 기본 구조


### while statement의 주의사항
while문은 반드시 종료 조건이 필요하다.
- 종료 조건이 없을 경우, 무한 반복에 빠져버린다.
- 따라서, False일 때 while문을 나갈 수 있게끔 False값을 유도한다.

```python
# while 반복문 예시
a = 0

while a < 5: # while문은 a >= 5가 되면 코드 실행을 멈춤
    print(a)
    a += 1
print('끝')
```

### 사용자 입력에 따른 반복
while문을 사용한 특정 입력 값에 대한 종료 조건 활용

```python
# 사용자 입력에 따른 반복문 예시
number = int(input('양의 정수를 입력하시오. : '))

while number <= 0:
    if number < 0:
        print(f'{number}는 음수입니다.')
        print()
    else:
        print('0은 양수, 음수로 정의되지 않습니다.')
        print()
    number = int(input('양의 정수를 다시 입력하시오. : '))

print(f'{number}는 양의 정수가 맞습니다.')
print('반복문을 종료합니다.')
```

### for과 while의 적절한 활용
for
- 반복 횟수가 명확하게 정해져 있는 경우에 유용
- 리스트, 튜플, 문자열 등과 같은 시퀀스 형식의 데이터, 딕셔너리, 세트 등과 같은 iter 데이터를 처리할 때

while
- 반복 횟수가 불명확하거나 조건에 따라 반복을 종료해야 할 때 유용
- 사용자의 입력을 받아 특정 조건이 충족될 때 까지(종료 조건에 해당) 반복하는 경우

### 반복 제어
반복문에서 때때로 일부만 실행해야 할 경우 사용하는 것

### 반복 제어의 종류
- break : 반복을 즉시 중지
- continue : 다음 반복으로 건너뜀

```python
# break 예시
number = int(input("양의 정수를 입력하시오. : "))
count = 0

while number <= 0:
    count += 1
    if count >= 4:
        print('프로그램을 종료합니다.')
        break
        
    if number < 0:
        print(f'{number}는 음수입니다.')
        
    else:
        print('0은 양수, 음수로 정의되지 않습니다.')
    number = int(input('양의 정수를 다시 입력하시오. : '))
print(f'{number}는 양수가 맞습니다!!!')
print(f'당신은 {count+1}번 만에 맞췄습니다.')
```

### 반복 제어 주의사항
- break, continue를 남용할 경우 코드의 가독성이 떨어짐
- 특정한 종료 조건을 만들거나,
  if문을 사용하는 등 여러가지 방법으로 둘을 대체할 수 있음
- 최대한 코드의 가독성을 유지하고 의도를 명확히 전달하는 것이 더 중요

### List Comprehension
간결하고 효율적인 리스트 생성법

### list comprehension의 구조


### if 조건문과의 연계
list comprehension 안에 if 조건문을 사용할 수 있다.


### List Comprehension 주의사항
간결하게 리스트를 표현할 수 있지만, 명시적이지는 않다.
- 남용 금지(simple is better than complex')

```python
# list comprehension 사용 예제
# 사용 전
numbers = [1, 2, 3]
squared_numbers = []
for number in numbers:
    number **= 2
    squared_numbers.append(number)
print(squared_numbers)

# 사용 후
numbers = [1, 2, 3]
squared_numbers = [number**2 for number in numbers]
print(squared_numbers)

# if 조건문 연계 예제
result = [i for i in range(10) if i % 2 == 0]
print(result)
```

## enumerate(iter, start = )
iter 객체의 각 요소를 인덱스와 함께 반환하는 파이썬 내장 함수
- start는 시작 index값을 설정해줌(없을 경우 디폴트는 0)

```python
alphabets = ['a', 'b', 'c', 'd']
for index, alphabet in enumerate(alphabets, start = 3): # 인덱스는 3, 4, 5, 6이 됨
    print(index, alphabet)
```
