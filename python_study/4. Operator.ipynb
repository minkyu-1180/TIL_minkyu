### 복합 연산자

- 연산과 할당이 함께 이루어짐

```python
# 복합 연산자 예시
y = 10 
y -= 4 # y에 y-4를 할당(연산 -> 할당)

z = 7
z *= 2 # z에 z*2를 할당

w = 15
w /= 3 # w에 z/3을 할당

q = 20
q //= 7 # w에 q//7을 할당

print(y)
print(z)
print(w)
print(q)
```

### 2. 비교 연산자

== : 같은 데이터값을 가졌는지 확인
- 동등성(equality)

is 비교 연산자 : 메모리 내에서 같은 객체를 참조하는지(주소) 확인
- None, True, False 등을 비교할 때 사용
- 식별성(identity)

```python
# is 연산 vs ==
# ==
print('Hi' == 'hi') # 둘은 다른 값 : False
print(2.0 == 2) # 둘은 같은 값 : True

# is
print(2.0 is 2) # 둘은 다른 객체를 참조 : False
# 파이썬 내부적으로도 이게 문법적인 문제는 없지만 보통 is이런식으로 안써서 ==를 쓰려하는게 맞나고 물어봄
```

### 3. 논리 연산자

- 비교 연산자와 함께 사용 가능

```python
# 논리 연산자 예제
print(True and False)
print(True or False)
print(not True)

# 논리 연산자 & 비교 연산자
num = 15
result = num > 10 and num % 2 == 0
print(result)

name = 'Minkyu'
age = 26
result = name == 'Alice' or age == 26
print(result)
```

### 단축평가
논리 연산에서 두 번째 피연산자를 평가하지 않고 경과를 결정하는 동작
- 코드 실행을 최적화하고, 불필요한 연산을 피할 수 있도록 함

1. and
- 첫 번째 피연산자가 False인 경우, 전체 표현식은 False로 결정(뒤의 피연산자 평가 X)
- 첫 번째 피연산자가 True인 경우, 전체 표현식의 결과는 두 번째 피연산자에 의해 결정
  두 번째 피연산자의 평가에 의해 표현식의 결과가 정해짐
2. or
- 첫 번째 피연산자가 True인 경우, 전체 표현식은 True로 결정(뒤의 피연산자 평가 X)
- 첫 번째 피연산자가 False인 경우, 전체 표현식의 결과는 두 번째 피연산자에 의해 결정
  두 번째 피연산자의 평가에 의해 표현식의 결과가 정해짐

```python
# 단축평가 예시
vowels = 'aeiou'

print(('a' and 'b') in vowels) 
'''
1. 'a' and 'b' : True and True라서 뒤의 True인 'b'가 결과값임
2. 'b' in vowels = False
'''
print(('b' and 'a') in vowels)
'''
1. 'b' and 'a' : True and True라서 뒤의 True인 'a'가 결과값임
2. 'a' in vowels = True
'''
print()
print(3 and 5)
'''
3 and 5 : True and True -> 뒤의 True인 5 출력
'''
print(3 and 0)
'''
3 and 0 : True and False -> False값인 0 출력
'''
print(0 and 3)
'''
0 and 3 : False and True -> False에서 그냥 계산 멈추고 0 출력
'''
print(0 and 0)
'''
0 and 0 : False and False -> 첫 False에서 그냥 계산 멈추고 0 출력
'''
print()
print(5 or 3)
'''
5 or 3 : False or False -> 뒤의 False값인 3 출력
'''
print(3 or 0)
'''
3 or 0 : True or False -> 첫 True에서 그냥 계산 멈추고 3 출력
'''
print(0 or 3)
'''
0 or 3 : False or True -> True값인 3 출력
'''
print(0 or 0)
'''
0 or 0 : False or False -> 뒤의 False값인 0 출력
'''
```

### 5. 시퀀스형 연산자

- 여기서 +와 *는 산술 연산자일 경우와 다른 역할
- 연산자 + : 두 시퀀스형 연산자를 연결
- 연산자 * : 시퀀스형 연산자를 정수 피연산자만큼 반복

시퀀스형 연산자 주의사항
- 시퀀스형 연산자로 연산 결과, 여러 피연산자 사이에 공백(space)가 없음 이거 주의하기
ex.
- print('My name is' + 'Minkyu') : My name isMinkyu
- print('My name is', 'Minkyu') : My name is Minkyu

```python
# 시퀀스형 연산자 예시
print('Hello' + 'Python')
print('hi' * 3)
print([1, 2] + [3, 4])
print([1, 2, 3] * 4)
```

### 여러 연산자들의 우선순위


1. 괄호
- 소괄호 grouping : () > 인덱싱, 슬라이싱 : []
2. 산술 연산자
- 거듭제곱(**) > 양수, 음수(+, -) > 산술1(*, /, //, %) > 산술2(+, -)
3. 비교 연산자
- 비교 : <, <=, >, >=, ==, !=
4. 객체 비교
- is, isnot
5. 멤버십 연산자
- in, not in
6. 논리 연산자
- not > and > or
