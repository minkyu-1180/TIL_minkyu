### 1. 모듈

### 개요
프로그램 전체를 개발자 혼자의 힘으로 작성하는 것은 드물다
- 이미 다른 프로그래머가 작성해놓은 여러 줄의 코드를 사용하는 것은 생산성에서 매우 중요!

### 모듈(Module)
한 파일로 묶인 변수와 함수의 모음
- 특정 기능을 하는 코드가 작성된 파이썬 파일(.py)

```python
# 모듈 예시
import math
print(math.pi)
```

### 2. 모듈의 활용
1. import module_name : 모듈 내의 요소들에 접근하기 전 모듈을 불러오는 단계
- from module_name import module_ele : 모듈에서 특정 요소를 불러온다.
    - 이후 해당 요소를 사용시 앞에 모듈.을 안붙여도 됨
    - 파이썬에서는 추천하지 않는 기능(서로 다른 모듈에서 같은 이름의 함수 제공시 문제 발생)
    - from module_name import *를 통해 한 번에 모든 요소를 import 가틍
  
2. help(module_name) : 내장함수 help를 통해 모듈의 기능 확인
모듈의 요소 : 모듈 내에 저장되어있는 변수, 함수, 클래스 등

3. module_name.module_ele : 모듈 내의 특정 요소를 사용
- dot(.) 연산자 의 의미 : 점의 왼쪽 객체에서 점의 오른쪽 이름을 찾아라!

### 사용자 정의 모듈
직접 정의한 모듈

### 사용자 정의 모듈 사용법
1. 모듈.py 작성
2. 해당 파이썬 확장 프로그램 안에 원하는 내용(변수, 함수, 클래스 등) 작성
3. 이후 다른 .py에서 해당 모듈 import 후 원하는 요소 사용

### 3. 파이썬 표준 라이브러리

### 파이썬 표준 라이브러리(Python Standart Library, PSL)
[PSL](https://docs.python.org/ko/3/library/index.html) : 파이썬 언어와 함께 제공되는 다양한 모듈과 패키지 모음

### 패키지(Package)
관련 모듈들을 하나의 디렉토리에 모아놓은 것

### 패키지 사용법
1. 디렉토리 생성
2. 해당 디렉토리에 여러가지 사용자 정의 모듈 작성
3. from package_name import module_name과 같은 방식으로 모듈 불러오기
- 만약, 패키지 안에 패키지가 있을 경우, from package_name_1.package_name_2 import module_name과 같은 방법으로 불러올 수 있음

### 패키지의 사용 목적
모듈들의 namespace을 구분하여 충돌을 방지하고, 효율적으로 관리하며 재사용할 수 있도록 돕기

### PSL 내부 패키지 vs 외부 패키지
PSL 내부 패키지 : 설치 없이 import 가능

외부 패키지 : 터미널에서 pip을 통해 설치 후 import(pip install out_pack_name)
- pip : 외부 패키지를 설치하는 것을 도와주는 파이썬 패키지 관리 시스템
- [PYPI](https://pypi.org)(Python Package Index)에 저장된 외부 패키지들을 설치해줌
- 설치 시 최신 버전, 특정 버전, 최소 버전 등을 명시하여 설치 가능
