# 7. Modules

## 1. Module

### Module
한 파일로 묶인 변수와 함수의 모음
- 특정한 기능을 위해 작성된 코들이 모인 파이썬 파일(.py)

#### 모듈 활용
1. 모듈 가져오기
   - import module_name을 통해 우리가 사용할 모듈을 불러올 수 있다.
   - 내장함수 help()를 사용하여 해당 모듈에 어떠한 기능이 들어있는지 볼 수 있다.
2. 모듈에 들어있는 요소 사용하기
   - module_name.method()를 통해 해당 모듈에 들어있는 요소인 method()의 기능을 사용할 수 있다.
   - dot(.) 연산자 : 왼쪽 객체에서 오른쪽 객체를 찾아라!
         - 처음 모듈을 불러올 때, from module_name import method를 할 경우, 불러온 이후 요소를 사용할 때 앞에 module_name.을 붙이지 않아도 된다.
         - from module_name import * : import module_name과 같은 기능(모든 요소 다 불러오기) / 다른 사용(앞에 안붙여도 됨)
         - 다른 모듈에서 같은 이름의 요소를 제공할 때 최근에 사용한 요소가 불려온다(이름 충돌 방지)
         - 명시적이지 않은 방법이기 때문에 그냥 import module_name을 한 후 가져다 쓰는 것이 더 좋은 방법

### 사용자 정의 모듈
직접 정의한 모듈
#### 사용자 정의 모듈 사용 방법
1. 모듈 작성(my_module.py)
2. 해당 모듈 안에 변수, 함수 등 작성
3. 다른 곳에서 해당 모듈을 불러온 후(import my_module), 작성한 요소들 호출(my_module.element())

## 2. 파이썬 표준 라이브러리

### 파이썬 표준 라이브러리(Python Standart Library, PSL)
[PSL](https://docs.python.org/ko/3/library/index.html) : 파이썬 언어와 함께 제공되는 다양한 모듈과 패키지들의 모음

### 패키지(Package)
관련 모듈들을 하나의 dir에 모아놓은 것
- PSL 내부 패키지 : 설치 없이 바로 import 할 수 있음
- 외부 패키지 : pip install 외부패키지명을 통해 터미널에서 설치 후 import 가능
      - [PYPI(Python Package Index)](https://pypi.org/) : 외부 패키지 저장소
      - 설치 시 최신 버전 / 특정 버전(==) / 최소 버전(>=)을 명시하여 설치 가능


#### 패키지 사용 목적
1. 모듈들의 이름공간을 구분하여 모듈 이름간의 충돌 방지
2. 비슷한 기능의 모듈들끼리 모아 효율적으로 관리하고 재사용할 수 있도록 보조

#### 패키지의 모듈 불러와서 사용하기 
, 
- 패키지 : my_inner_pack, my_pack_1, my_pack_2
      - my_pack_1, my_pack_2는 my_inner_pack의 하위 패키지
- 모듈 : my_module_1.py, my_module_2.py

## + 새롭게 알게 된 부분
- 기존에 numpy, pandas, matplotlib 등 머신러닝에 필요한 모듈들을 불러온 경험이 있었는데, 이 과정에서 .의 역할에 대해서는 새롭게 알게 되었다.
- 내가 직접 모듈을 작성하고, 여러 모듈들을 패키지에 담아보니 사용자 정의 모듈 뿐만 아니라 파이썬에서 지원하는 모듈들을 불러오는 과정에 대해 더 확실하게 깨달았다. 특히, 내가 사용하고 있는 파이썬 확장 파일이 해당 패키지들와 동일한 선상에 있어야 한다는 점을 알게 되었다.
