```python
#  -  
original = [1, 2, 3]
copy = original #   
copy[0] = 100 #    

#    ->        
print(original) # [100, 2, 3]
print(copy) # [100, 2, 3]
```

####  (Shallow Copy)
  


-       
```python
original = [1, 2, 3]
copy = original[:] #     ->   

copy[0] = 50 #      ->   X

print(original) # [1, 2, 3]
print(copy) # [50, 2, 3]
```

####  (Deep Copy)
copy  deepcopy   


-         
```python
import copy # copy  

original = [1, 2, [1, 2, 3]]
deep_copy = copy.deepcopy(original) # deepcopy    

deep_copy[2][1] = 50
print(original) # [1, 2, [1, 50, 3]]
print(deep_copy) # [1, 2, [1, 50, 3]]
```
