# 9. Data Structure

## 1. Data Structure

### Data Structure
여러 데이터를 효과적으로 사용, 관리하기 위한 구조
- str, list, dict 등이 데이터 구조이다.

#### 자료구조
컴퓨터 공학에서 Data Structure을 지칭하는 단어


#### 데이터 구조 활용
각 데이터 구조의 '메서드'를 호출하여 다양한 기능을 활용

### 메서드(method)
객체에 속한 함수
- 객체의 상태를 조작, 동작 수행
- 클래스 내부에서 정의되는 함수
- 각 Data type 별로 다양한 기능을 가진 메서드 존재

#### 메서드 호출 방법
Object.method() : 데이터 타입(클래스)의 객체가 해당 클래스에서 정의된 메서드(함수)를 호출한다.

#### 메서드의 기능
메서드는 여러 개를 dot(.)으로 이어서 순차적으로 적용 가능(chained)
- 앞의 메서드 적용 결과가 뒤의 메서드를 적용할 수 있어야 사용 가능


## 2. Sequence Data Structure

### 문자열

#### 문자열의 조회/탐색/검증 메서드
1. s.find(x) : x의 첫 번째 위치를 반환
   - 없으면 -1 반환
2. s.index(x) : x의 첫 번째 위치를 반환
   - 없으면 오류 발생(ValueError : substring not found)
3. s.isupper() : 문자열 s가 모두 대문자로 이루어져 있는지 확인(True / False)
4. s.islower() : 문자열 s가 모두 소문자로 이루어져 있는지 확인(True / False)
5. s.isalpha() : 문자열 s가 모두 영문 알파벳으로 이루어져 있는지 확인(True / False)
6. s.istitle() : 문자열 s 내의 공백 바로 뒤에 오는 문자가 대문자인지 확인(True / False)
   - 해당 문자를 제외한 나머지 문자들은 소문자여야 True 반환

#### 문자열의 조작 메서드(새 문자열 반환)
- []로 표기된 부분은 선택인자를 의미
- 해당 메서드를 사용할 경우, 원본을 변환하는 것이 아닌 새로운 복제본을 만들어냄(원본의 변화 X)

1. s.replace(old, new[, count]) : 바꿀 대상 글자(old)를 새로운 글자(new)로 바꿔서 반환
2. s.strip([chars]) : 문자열 양쪽의 공백이나 특정 문자 제거
   - 제거하려는 특정 문자가 해당 문자열 양쪽에 없을 경우, 복사본 반환
3. s.split(sep = None, maxsplit = -1) : 공백이나 특정 문자를 기준으로 분리하여 문자열의 리스트 반환
   - maxsplit : 문자열을 몇 번 분리할 것인지(ex. sep = '.', maxsplit = 3 -> 앞의 3개의 '.'를 기준으로 분리하여 리스트 반환)
4. 'separater'.join([iter]) : separator을 기준으로 나누어진 iterable 데이터를 하나의 문자열로 합쳐서 반환
5. s.capitalize() : 가장 첫 번째 글자를 대문자로 변경
6. s.title() : 문자열 내에서 띄어쓰기를 기준으로 각 단어의 첫 글자는 대문자로, 나머지는 소문자로 변환
7. s.upper() : 문자열의 영문 알파벳을 모두 대문자로 변경
8. s.lower() : 문자열의 영문 알파벳을 모두 소문자로 변경
9. s.swapcase() : 문자열의 대문자는 소문자로, 소문자는 대문자로 변경

### 리스트

#### 리스트의 값 추가/삭제 메서드
- 복사본이 없이 원본에 적용 -> 메서드를 실행한 수 원본을 출력해야 적용된 값이 나옴
1. l.append(x) : 리스트의 마지막 항목으로 x을 추가
2. l.extend(iter) : 리스트의 끝에 iter 객체를 추가
   - l = l += iter의 기능
3. l.insert(i, x) : 리스트의 i번 index에 x을 삽입
   - 기존의 i ~ N-1번 index의 요소들이 i+1 ~ N번 index의 요소들로 바뀜(한 칸씩 밀림)
4. l.remove(x) : 리스트의 가장 왼쪽에 있는 x 제거
   - x가 리스트의 요소가 아닐 경우 ValueError 발생
5. l.pop() : 리스트의 마지막 요소 반환 후 제거
   - 제거된 마지막 요소는 반환되기 때문에 해당 값을 새로운 변수에 할당 가능
   - l.pop(i) : 리스트의 i번 index 요소 반환 후 제거
6. l.clear() : 리스트의 모든 항목 삭제
   - l.clear()을 출력할 경우, None이 출력됨!!! 이후 l을 출력하면 빈리스트 출력
  
#### 리스트의 탐색/정렬 메서드
1. l.index(x, start, end) : 리스트에 있는 요소 중 가장 작은 index에 있는 x의 index값 반환
   - start, end 사이(디폴트 : 0 ~ len(l)-1)에 없을 경우 ValueError 발생
2. l.reverse() : 원본 리스트의 순서를 역순으로 변경
3. l.sort() : 원본 리스트를 오름차순으로 정렬
   - 디폴트 : reverse = False
   - 내림차순 : reverse = True
   - 리스트 내에 다른 data type의 요소들이 존재할 경우 TypeError 발생
   - 오름차순일 때 특수문자 - 대문자 - 소문자 순으로 정렬됨
   - vs sorted() : sorted()함수는 복사본을 반환 / .sort()메서드는 원본에 적용
4. l.count(x) : 리스트의 요소들 중 x의 개수 반환
   - 없을 경우 0 반환

### 문자열에 포함된 문자들의 유형을 판별하는 메서드

1. isdecimal() : 문자열이 모두 숫자 문자 (0 ~ 9)로만 이루어져 있어야 True
2. isdigit() : 유니코드 숫자까지 인식해서 True
3. isnumeric() : 몇 가지 추가적인 유니코드(분수, 지수, 루트 등) 문자까지 인식해서 True 

## 3. Non-sequence Data Structure

### 세트

#### 세트의 메서드
1. s.add(x) : 집합에 요소 x를 추가
   - 기존 집합에 x 요소가 존재했을 경우 변화 X
2. s.clear() : 집합의 모든 요소 제거
3. s.remove(x) : 집합에서 x 요소 제거
   - 기존 x가 집합에 없을 경우 Key Error 발생
4. s.pop() : 집합에서 랜덤하게(순서 X) 요소 반환 후 해당 요소 제거
5. s.discard(x) : 집합에서 x 요소 제거
   - 기존 x가 집합에 없을 경우 집합 내에 변화 X
6. s.update(iter) : 집합에 iterable 요소 추가
   - iter 요소가 아닐 경우 TypeError 발생
#### 세트의 집합 메서드
1. s1.difference(s2) : s1 - s1
2. s1.intersection(s2) : s1 & s2
3. s1.union(s2) : s1 | s2
4. s1.issubset(s2) : s1 <= s2이면 True 반환
5. s1.issuperset(s2) : s1 >= s2이면 True 반환

### 딕셔너리

#### 딕셔너리 메서드
1. d.clear() : 딕셔너리의 모든 키-값 쌍 요소 제거
2. d.get(k[, default]) : 키에 연결된 값을 반환
   - 해당 키 k가 딕셔너리에 존재하지 않을 시 None 혹은 default값 반환
   - vs d[key] : d[key]같은 경우, 해당 키가 존재하지 않을 시 Error 발생
   - 코드 실행 흐름을 살리고 싶은 경우 get() 메서드 사용
3. d.keys() : 딕셔너리의 모든 키를 모은 객체 반환(반환 타입 : dict_keys(['key1', 'key2', 'key3', ,,,]))
   - 이건 리스트형이 아니고, 그냥 딕셔너리 키 타입임
4. d.values() : 딕셔너리의 모든 값들을 모은 객체 반환(반환 타입 : dict_values(['val1', 'val2', 'val3', ,,,]))
   - 딕셔너리 값 타입
5. d.items() : 딕셔너리의 모든 키 - 값 쌍을 모은 객체 반환(반환 타입 : dict_items([('key1', 'val1'), ,,, , ]))
   - 딕셔너리 아이템 타입
6. d.pop(k[, default]) : d에서 키를 제거하고 연결되어 있던 값을 반환
   - 해당 키가 없을 시 Error가 발생하거나 default값을 반환
7. d.setdefault(k[, default]) : d에서 키와 연결된 값을 반환
   - 해당 키가 없을 시, k : default 쌍을 딕셔너리에 추가한 후 default값 반환
8. d.update([other]) : 다른 딕셔너리 내의 각 키에 대해 업데이트 시행(덮어쓰기)
   - 기존에 d에 있던 키인 경우, 해당 키의 대응 값을 other에 있는 값으로 대처
   - 기존에 d에 없던 키인 경우, 새로운 키 - 값 쌍을 d에 추가

## 4. 복사

### 복사
복사는 데이터 타입에 따라 달라진다.
- 변경 가능한 데이터 타입과 변경 불가능한 데이터 타입을 다른 방식으로 다룬다

#### 변경 가능한 데이터 타입의 복사
리스트 : 복사 후 요소를 수정해도 같이 수정됨
- 리스트 요소의 경우 같은 객체를 참조
```python
a = [1, 2, 3, 4]
b = a # 복사
b[0] = 100
print(a) # [100, 2, 3, 4]
print(b) # [100, 2, 3, 4]
```

#### 변경 불가능한 데이터 타입의 복사
문자열 : 복사할 경우 서로 다른 메모리주소를 저장
- 수정 시 참조하는 객체가 달라짐
```python
a = 10
b = a
b = 20
print(a) # 10
print(b) # 20
```

### 복사의 유형

#### 할당(Assignment)
할당 연산자를 통한 복사


- 해당 객체에 대한 객체 참조를 복사하는 것
```python
# 할당 - 리스트 복사
original = [1, 2, 3]
copy = original # 할당을 통해 복사
copy[0] = 100 # 복사본의 인덱스 값 수정

# 객체를 참조하는 것 -> 요소가 수정되면 바뀐 객체의 주소를 같이 바라봄 
print(original) # [100, 2, 3]
print(copy) # [100, 2, 3]
```

#### 얕은 복사(Shallow Copy)
슬라이싱을 통한 복사


- 슬라이싱을 통해 생성된 객체는 원본 객체와 독립
```python
original = [1, 2, 3]
copy = original[:] # 슬라이싱을 통한 얕은 복사 -> 독립적인 객체 생성

copy[0] = 50 # 다른 객체의 인덱스 객체 수정 -> 원본에 영향 X

print(original) # [1, 2, 3]
print(copy) # [50, 2, 3]
```
#### 얕은 복사의 한계


변경 가능한 객체 안에 변경 가능한 객체가 있는 경우(ex. 2차원 리스트), 원본과 복사본의 주소는 다르다 하지만!!!
- 내부 객체의 주소는 같다 -> 수정 시 함께 변경됨

#### 깊은 복사(Deep Copy)
copy 모듈이 deepcopy 메서드를 사용한 복사


- 내부에 중첩된 모든 객체까지 새로운 객체 주소를 참조하게 함
```python
import copy # copy 모듈 불러오기

original = [1, 2, [1, 2, 3]]
deep_copy = copy.deepcopy(original) # deepcopy 메소드를 활용한 깊은 복사

deep_copy[2][1] = 50
print(original) # [1, 2, [1, 50, 3]]
print(deep_copy) # [1, 2, [1, 50, 3]]
```
