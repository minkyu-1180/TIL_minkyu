## 3. 비교 연산자

### 비교 연산자
연산의 결과를 bool형이 나오게 하는 연산자

#### 비교 연산자의 종류


### is 비교 연산자
메모리 내에서 같은 객체를 참조하는지 확인

#### ==과 is의 차이
- == : 동등성(equality) 비교 - 같은 값(데이터)인가?
- is : 식별성(identity) 비교 - 메모리 내에서 같은 객체(주소)를 참조하는가?

#### is 비교 연산자의 활용
None, True, False 등을 비교시 사용

## 4. 논리 연산자

#### 논리 연산자의 종류



### 단축평가
논리 연산에서 두 번째 피연산자를 평가하지 않고 결과를 결정하는 동작
#### 단축평가의 필요성
코드 실행을 최적화하고, 불필요한 연산이 줄어듦

#### 단축평가 Case
1. and Case
    - 첫 번째 피연산자가 False인 경우(False and ???) : 전체 표현식은 첫 번째 피연산자 False로 결정(단축평가가 이뤄짐)
    - 첫 번째 피연산자가 True인 경우(True and ???) : 전체 표현식은 두 번째 피연산자의 평가 결과
2. or Case
   - 첫 번째 피연산자가 True인 경우(True or ???) : 전체 표현식은 첫 번째 피연산자 True로 결정(단축평가가 이뤄짐)
   - 첫 번째 피연산자가 False인 경우(False or ???) : 전체 표현식은 두 번째 피연산자의 평가 결과

## 7. 연산자 간 우선순위

## + 새롭게 알게 된 내용
- 비교 연산자 중 데이터 값 자체가 같은지를 물어보는 연산자(==)와 두 객체가 저장된 메모리 주소가 같은지를 물어보는 연산자(is)의 차이점을 좀 더 자세히 배울 수 있었다. 참고로, id()라는 내장함수를 사용하여 해당 변수가 참조하는 객체의 메모리주소를 볼 수 있는데, 이를 이용할 수도 있다.
- 단축평가라는 개념은 몰라도 실행해보며 어렴풋이 알고는 있었는데, 단축평가 과정에서 (True and True) 혹은 (False or False)같이 전체 결과가 뒤의 평가 결과에 의해 정해지는 경우, 전체 결과 자체가 뒤의 평가 결과로 대체된다는 것을 알게 되었다. 초반에는 조금 헷갈렸는데, 몇 가지의 예시를 들고 나니 확실히 이해가 되었다.
- 연산자 간 우선순위는 아직 봐도 헷갈린다. 다행히 같은 연산 케이스끼리 묶여있는 경우가 많기 때문에, 논리 구성을 따라가다 보면 어느새 자연스레 연산 순서를 알게 될 것 같다.
