# 3. Type Conversion

## 1. Type Conversion

### Type Conversion의 종류

#### Implicit Type Conversion
암시적 형 변환 : 파이썬이 자동으로 형 변환을 실시하는 것
- Boolean과 Numeric Type에서만 가능
- int -> float / bool -> int 등

```python
print(8 + 1.1) # 8.0 + 1.1 = 9.1
print(True + 1) # 1 + 1 = 2
print(True - False) # 1 - 0 = 0
```

#### Explicit Type Conversion
명시적 형 변환 : 개발자가 직접 형 변환을 실시하는 것
- 암시적 형 변환을 제외한 모든 형 변환
- str -> int : 형식에 맞는 숫자만 가능
- int -> str : 모든 case 다 가능

```python
print(int('1')) # str(1) -> int(str(1))
print(str(1) + str(0)) # '1' + '0' = '10'
print(float('3.5')) # 3.5

# ValueError: invalid literal for int() with base 10: '3.5' 
# print(int('3.5')) 

print(int(3.5)) # 3
```

## 2. Collection 간의 형 변환

## + 새롭게 알게 된 내용
- 우선 형 변환은 알고 있었는데, 파이썬에서 자체적으로 형 변환을 해주는 암시적 형 변환에 대해서는 처음 알게 되었다. 몇 가지 특이 케이스에 한해서 진행되는 형 변환이다.
- int('3.4')처럼 str형의 float값을 다시 int형으로 바꾸는 과정에서 오류가 발생했다. 오류 내용은 ValueError: invalid literal for int() with base 10: '3.5' 인데, 형 변환 과정에서 int() 내장함수는 '정수 문자열' 또는 '실수'만 인자로 받는다는 것이 그 이유였다. 위의 예시처럼 '3.4'는 '실수 문자열'이기 때문에 오류가 발생하는 것이었다. 해결 방법은 우리가 중간에 생략한 float() 형 변환을 넣어주면 된다.
- dictionary의 요소들은 key-value 쌍으로 이루어져 있어 당연히 형 변환이 안될 것이라 생각했는데, list, tuple, set에 한해서만 value값들을 요소로 하게끔 가능했다. 이를 이용하여 우리가 필요한 값들이 dict의 value에 있을 경우, 형 변환을 통해 추출하기 용이할 것 같다.
